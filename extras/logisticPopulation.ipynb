```python
import matplotlib.pyplot as plt
import random
# Because this is a recursive function, we need to write custom tokenizer scheme.

def logistic_map(r, x):
    noise=random.uniform(-1,1)*0.001
    noise=0
    return r * x * (1 - x) + noise

num_train=15
num_test=35
total_num=num_train+num_test
r=1.2

data_train=[0.01]
for _ in range(num_train):
    data_train.append(logistic_map(r, data_train[-1]))

data_test=[data_train[-1]]
for _ in range(num_test):
    data_test.append(logistic_map(r, data_test[-1]))


plt.plot(data_train, label='Training Data')
plt.plot(range(num_train, total_num+1), data_test, label='Testing Data')
plt.title('Population Dynamics in the Logistic Map')
plt.xlabel('Generation')
plt.ylabel('Population')
plt.legend()
plt.show()

```

```python
#get the min and max of the test and train data
min_data=min(data_train+data_test)
max_data=max(data_train+data_test)

#data already lies between 0 and 1. Multiply by 999 to get the range between 0 and 999
data_train=[int(999*(x-min_data)/(max_data-min_data)) for x in data_train]
data_test=[int(999*(x-min_data)/(max_data-min_data)) for x in data_test]

plt.plot(data_train, label='Training Data')
plt.plot(range(num_train, total_num+1), data_test, label='Testing Data')
plt.title('Population Dynamics in the Logistic Map')
plt.xlabel('Generation')
plt.ylabel('Population')
plt.legend()
plt.show()

print(data_train)
print(data_test)
```
