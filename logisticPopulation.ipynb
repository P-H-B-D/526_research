```python
import matplotlib.pyplot as plt

# Because this is a recursive function, we need to write custom tokenizer scheme.

def logistic_map(r, x):
    return r * x * (1 - x)

# Values of r for both scenarios
r_vals=[0.75,0.85,0.95,1.2,1.5,2.5]
iterations = 30
x = 0.01

for r in r_vals:
    population_die_out = [x]
    for _ in range(iterations):
        population_die_out.append(logistic_map(r, population_die_out[-1]))

    plt.plot(population_die_out, label=f'r = {r}')

plt.title('Population Dynamics in the Logistic Map for Different Scenarios')
plt.xlabel('Generation')
plt.ylabel('Population')
plt.legend()
plt.show()
```
